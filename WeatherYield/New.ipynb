```python
# Read RMA county yields report table
rma_df = spark.table("workspace.default.rma_county_yields_report_399")

# Read NOAA station monthly metrics table
noaa_df = spark.table("workspace.default.noaa_station_month_metrics_399")

# Join on Year and State columns
combined_df = rma_df.join(
    noaa_df,
    (rma_df["Yield Year"] == noaa_df["Year"]) & (rma_df["State Name"] == noaa_df["State Name"]),
    "inner"
)

display(combined_df)
```
